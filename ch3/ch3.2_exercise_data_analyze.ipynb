이제 2장에서 했던 것처럼 pandas로 읽어보자.

````python
import pandas as pd
not_exercise = pd.read_excel("./data/reason_not_exercise.xls", skiprows=1)
not_exercise.head()
````

이 데이터는 다양한 분류로 데이터를 구분하고 있다. 그러나 우리는 '기간'이라는 칼럼은 의미가 없다.  
그러므로 이 칼럼은 삭제하겠다. 그전에 not_exercise.head()의 ehad()는 앞부분 5개만 보여달라는 의미이다.

````python
not_exercise.drop(columns='시점', inplace=True)
not_exercise
````

그리고 서울시 구별 데이터는 지금의 목적과는 관계가 없어서 인덱스 28부터는 지우겠다.

````python
not_exercise.drop(index=range(22, 52), inplace=True)
not_exercise
````

이렇게 정리한 후 대분류 또는 분류에서 데이터를 뽑아 그래프로 확인해 보자.

````python
not_exercise['구분별(1)'] == '성별'
````

대분류 칼럼에서 '성별'만 찾으면 위와 같은 결과가 나온다.  
이 조건문을 이용해서 대분류가 성별인 행만 추려낼 수 있다.

````python
not_ex_sex = not_exercise[not_exercise['구분별(1)']=='성별'].copy()
not_ex_sex
````

여기서 잠시 파이썬의 copy 규칙을 들여다볼 필요가 있다.

````python
test_a = [1, 2, 3, 4]
test_a
````

test_a를 [1, 2, 3, 4]로 저장한다.

````python
test_b = test_a
test_b
````

그리고 test_b에 test_a를 카피(copy)한다.

````python
test_b[2] = 0
test_b
````

그리고 test_b의 2번째(실제로는 3번째)를 0으로 변경한다.  
그럼 당연히 변경이 될 것이다. test_b를 변경 후 test_a를 확인해보자.

````python
test_a
````

test_b를 바꾸었더니 test_a를 바꾸었다.
```
test_b = test_a
```
라고 할 때, test_b를 위한 공간이 준비되는 것이 아니다.   
```
[1, 2, 3, 4]
```
라는 데이터가 저장된 곳 하나를 test_a와 test_b가 모두 가리키고 있는 것이다.  
그래서 test_b의 내용을 바꾸면 test_a도 바뀌게 된다.

````python
test_c = test_a.copy()
test_c
````

그래서 등호(=)를 이용해서 변수를 저장할 때는 copy() 옵션을 사용한다.

````python
test_c[2] = 100
test_c
````

````python
test_a
````

그러면 위 코드처럼 변경되지 않는다.  
다시 본래 코드로 돌아와서 이 상태에서 그래프를 조금 간결하게 그리기 위해 대분류(구분별(1)) 칼럼을 지우겠다.

````python
not_ex_sex.drop(columns='구분별(1)', inplace=True)
not_ex_sex
````

여기서 그래프를 조금 편하게 그리기 위해 '구분별(2)'를 인덱스로 바꾸겠다.

````python
not_ex_sex.set_index('구분별(2)', inplace=True)
not_ex_sex
````

````python
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

# f_path = "/Library/Fonts/AppleGothic.ttf"
f_path = "C:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font', family=font_name)
````

````python
f, ax = plt.subplots(1, 2, figsize=(16, 8))

not_ex_sex['운동을 할 충분한 시간이 없어서'].plot.pie(explode=[0, 0.02], ax=ax[0], autopct='%1.1f%%')
ax[0].set_title('운동을 할 충분한 시간이 없어서')
ax[0].set_ylabel('')

not_ex_sex['함께 운동을 할 사람이 없어서'].plot.pie(explode=[0, 0.02], ax=ax[1], autopct='%1.1f%%')
ax[1].set_title('함께 운동을 할 사람이 없어서')
ax[1].set_ylabel('')

plt.show()
````

이제 그래프를 보면 약간 미묘하게 남/여의 차이를 알 수 있다.  
근소한 차이이긴 하지만, 남자들은 운동할 충분한 시간이 없다고 답변하고, 여성들은 함께 운동할 사람이 없어서라고 답변을 했다는 것입니다.

````python
not_ex_age = not_exercise[not_exercise['구분별(1)']=='연령별'].copy()
not_ex_age.drop(columns='구분별(1)', inplace=True)
not_ex_age.set_index('구분별(2)', inplace=True)
not_ex_age
````

이제는 성별로 그릴 때 했던 코드를 한 번에 다 적용하고 위 코드처럼 데이터를 얻겠다.

````python
f, ax = plt.subplots(1, 3, figsize=(16, 8))

explode_setting = [0.02, 0.02, 0.02, 0.02, 0.02, 0.02]

not_ex_age['운동을 할 충분한 시간이 없어서'].plot.pie(explode=explode_setting, ax=ax[0], autopct='%1.1f%%')
ax[0].set_title('운동을 할 충분한 시간이 없어서')
ax[0].set_ylabel('')

not_ex_age['함께 운동을 할 사람이 없어서'].plot.pie(explode=explode_setting, ax=ax[1], autopct='%1.1f%%')
ax[1].set_title('함께 운동을 할 사람이 없어서')
ax[1].set_ylabel('')

not_ex_age['운동을 할 만한 장소가 없어서'].plot.pie(explode=explode_setting, ax=ax[2], autopct='%1.1f%%')
ax[2].set_title('운동을 할 만한 장소가 없어서')
ax[2].set_ylabel('')

plt.show()
````

그리고 다시 그래프를 보면, 나이가 많을수록 함께 운동할 사람이 없어서라는 대답이 많아지고, 나이가 어릴 수록 시간이 없다는 이유를 말하는 경우가 보인다.
